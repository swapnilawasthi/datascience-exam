### Q 1.1: Fibionacci
![fib image](https://upload.wikimedia.org/wikipedia/commons/thumb/9/93/Fibonacci_spiral_34.svg/200px-Fibonacci_spiral_34.svg.png)

#### Q 1.1.1
Given $n$ where $n \in \mathbb{N}$ (i.e., $n$ is an integer and $n > 0$), write a function `fibonacci(n)` that computes the Fibonacci number $F_n$, where $F_n$ is defined by the recurrence relation:

$$ F_n = F_{n-1} + F_{n-2}$$

with initial conditions of:

$$ F_1 = 1,  F_2 = 1$$

```python
import sys

fibDict = {1:1, 2:1}

def fibonacci(n):
    if n <= 0:
        return 0
    if n == 1 or n == 2:
        return fibDict[n]
    if n not in fibDict:
        fibDict[n] = fibonacci(n-1) + fibonacci(n-2)
        return fibonacci(n-1) + fibonacci(n-2)
    else:
        return fibDict[n]
```

#### Q 1.1.2
What's the complexity of your implementation?

The complexity of this implementation is O(n) for the first time(s) the function is run. When it is run numerous times, the complexity would be O(m+q), where m is the maximum of the n's that the function is run on and q is the number of queries. This function also uses O(n) of memory, since the computed Fibonacci’s are stored.

### Q 2.1: Finding $\pi$ in a random uniform?
![pi pie img](http://core2.staticworld.net/images/article/2016/03/pi-day-intro-100649273-carousel.idge.jpeg)

Given a uniform random generator $[0,1)$ (e.g., use your language's standard libary to generate random value), write a a function `compute_pi` to compute [$\pi$](https://en.wikipedia.org/wiki/Pi).

```python
import sys
import random
import math

def compute_pi(numPoints):
    if numPoints <= 0:
        return 0
    inCircle = 0

    for i in range(numPoints):
        xCord = random.uniform(0,1)
        yCord = random.uniform(0,1)

        if math.hypot(xCord, yCord) <= 1:
            inCircle += 1
    return (4.0 * inCircle) / numPoints
```

For this implementation, the higher number that is used as the parameter, the more accurate the pi estimation will be. The way this works is by generating a large number of random coordinates and then calculating the norm between the two points. If the length of this vector is greater than 1, then that point would not be inside a quarter of an imaginary circle or diameter size 2. Once all of the points that are in the circle are found, we multiply these by four to get how many would be inside a full circle and devide this by the total number of generated points.
